Problem-2: Array Manipulation

```python
import numpy as np

# Task 1: Create an array with values ranging from 10 to 49
array_10_49 = np.arange(10, 50)
print("Task 1 - Array 10 to 49:\n", array_10_49)

# Task 2: Create a 3x3 matrix with values ranging from 0 to 8
matrix_3x3 = np.arange(9).reshape((3,3))
print("\nTask 2 - 3x3 Matrix 0 to 8:\n", matrix_3x3)

# Task 3: Create a 3x3 identity matrix
identity_matrix = np.eye(3)
print("\nTask 3 - 3x3 Identity Matrix:\n", identity_matrix)

# Task 4: Create a random array of size 30 and find its mean
random_array_30 = np.random.random(30)
mean_value = random_array_30.mean()
print("\nTask 4 - Random Array 30 Elements:\n", random_array_30)
print("Mean of Random Array:", mean_value)

# Task 5: Create a 10x10 array with random values and find min and max
random_array_10x10 = np.random.random((10,10))
min_value = random_array_10x10.min()
max_value = random_array_10x10.max()
print("\nTask 5 - 10x10 Random Array:\n", random_array_10x10)
print("Minimum Value:", min_value)
print("Maximum Value:", max_value)

# Task 6: Create a zero array of size 10 and replace 5th element with 1
zero_array = np.zeros(10)
zero_array[4] = 1   # Indexing starts at 0
print("\nTask 6 - Zero Array with 5th Element as 1:\n", zero_array)

# Task 7: Reverse an array arr = [1,2,0,0,4,0]
arr = np.array([1,2,0,0,4,0])
reversed_arr = arr[::-1]
print("\nTask 7 - Reversed Array:\n", reversed_arr)

# Task 8: Create a 2D array with 1 on the border and 0 inside
n = 5  # Example size 5x5
border_array = np.ones((n,n))
border_array[1:-1, 1:-1] = 0
print("\nTask 8 - 2D Array with Border 1s and Inside 0s:\n", border_array)

# Task 9: Create an 8x8 matrix and fill it with a checkerboard pattern
checkerboard = np.zeros((8,8), dtype=int)
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("\nTask 9 - 8x8 Checkerboard Pattern:\n", checkerboard)
```

Problem-3: Array Operations

```python
import numpy as np

# Given arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# Task 1: Add the two arrays
add_xy = x + y
print("Task 1 - x + y:\n", add_xy)

# Task 2: Subtract the two arrays
sub_xy = x - y
print("\nTask 2 - x - y:\n", sub_xy)

# Task 3: Multiply the array with an integer (let's choose 3)
mul_x = x * 3
print("\nTask 3 - x * 3:\n", mul_x)

# Task 4: Square of each element
square_x = x**2
print("\nTask 4 - Square of x:\n", square_x)

# Task 5: Dot product
dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)
print("\nTask 5 - Dot Product:")
print("v . w =", dot_vw)
print("x . v =", dot_xv)
print("x . y =\n", dot_xy)

# Task 6: Concatenate x and y along row (axis=0) and v and w along column (axis=1)
concat_xy_row = np.concatenate((x, y), axis=0)
concat_vw_col = np.vstack((v, w)).T  # transpose to make column
print("\nTask 6 - Concatenation:")
print("x and y along row:\n", concat_xy_row)
print("v and w along column:\n", concat_vw_col)

# Task 7: Concatenate x and v
try:
    concat_xv = np.concatenate((x, v), axis=0)
except ValueError as e:
    print("\nTask 7 - Concatenate x and v Error:\n", e)
    print("Explanation: Cannot concatenate x (2x2) with v (1x2) along axis=0 because the number of columns must match. For axis=1, number of rows must match. Shapes must align.")

```

Problem 4: Matrix Operations

```python
import numpy as np

# Given matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Task 1: Prove A * A_inv = I
A_inv = np.linalg.inv(A)  # Inverse of A
identity_check = np.dot(A, A_inv)
print("Task 1 - A * A_inv:\n", identity_check)

# Task 2: Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nTask 2 - AB:\n", AB)
print("BA:\n", BA)
print("AB == BA?", np.array_equal(AB, BA))

# Task 3: Prove (AB)^T = B^T * A^T
AB_T = np.dot(A, B).T
B_TA_T = np.dot(B.T, A.T)
print("\nTask 3 - (AB)^T:\n", AB_T)
print("B^T * A^T:\n", B_TA_T)
print("Check equality:", np.array_equal(AB_T, B_TA_T))

# Task 4: Solve system of linear equations using matrix inverse
# Equations:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

# Represent in matrix form AX = B
A_matrix = np.array([[2, -3, 1],
                     [1, -1, 2],
                     [3, 1, -1]])
B_matrix = np.array([-1, -3, 9])

# Solve using inverse
A_matrix_inv = np.linalg.inv(A_matrix)
X_solution = np.dot(A_matrix_inv, B_matrix)
print("\nTask 4 - Solution using inverse method:\n", X_solution)

# Optional: Solve using np.linalg.solve (more efficient)
X_solution_solve = np.linalg.solve(A_matrix, B_matrix)
print("Solution using np.linalg.solve:\n", X_solution_solve)
```
