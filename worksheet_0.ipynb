```python
# Create empty lists
low_study = []
moderate_study = []
high_study = []

# Classify study hours
for record in time_data:
    study_hours = record[0]
    if study_hours < 3:
        low_study.append(study_hours)
    elif 3 <= study_hours <= 5:
        moderate_study.append(study_hours)
    else:
        high_study.append(study_hours)

# Print the lists
print("Low Study Hours:", low_study)
print("Moderate Study Hours:", moderate_study)
print("High Study Hours:", high_study)
```

Task 1: Sum of Nested Lists

```python
def sum_nested_list(nested_list):
    """
    Recursively sum all numbers in a nested list.
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):  # If element is a list, recurse
            total += sum_nested_list(element)
        else:  # If element is a number, add to total
            total += element
    return total

# Test
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print("Sum of nested list:", sum_nested_list(nested_list))
```

Task 2: Generate All Permutations of a String

```python
def generate_permutations(s):
    """
    Recursively generate all permutations of a string s.
    """
    if len(s) == 0:
        return ['']

    permutations = []
    for i, char in enumerate(s):
        # Remaining string without the current character
        remaining = s[:i] + s[i+1:]
        # Recursively get permutations of the remaining string
        for perm in generate_permutations(remaining):
            permutations.append(char + perm)

    # Return unique permutations only
    return list(set(permutations))

# Test
print("Permutations of 'abc':", generate_permutations("abc"))
print("Permutations of 'aab':", generate_permutations("aab"))
```

Task 3: Directory Size Calculation

```python
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}
```

```python
def calculate_directory_size(directory):
    """
    Recursively calculate the total size of files in a directory.
    """
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):  # Subdirectory
            total_size += calculate_directory_size(value)
        else:  # File size
            total_size += value
    return total_size

# Test
print("Total directory size:", calculate_directory_size(directory_structure), "KB")
```

 Task 1- Coin Change Problem:


```python
def min_coins(coins, amount):
    """
    Find the minimum number of coins to make up a given amount using DP.
    """
    dp = [float('inf')] * (amount + 1)  # dp[i] = min coins to make amount i
    dp[0] = 0  # 0 coins needed to make amount 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

coins = [1, 2, 5]
amount = 11
print("Minimum coins needed:", min_coins(coins, amount))
```

Task 2: Longest Common Subsequence (LCS)

```python
def longest_common_subsequence(s1, s2):
    """
    Returns the length of the longest common subsequence between s1 and s2.
    """
    m, n = len(s1), len(s2)
    dp = [[0]*(n+1) for _ in range(m+1)]

    for i in range(1, m+1):
        for j in range(1, n+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = 1 + dp[i-1][j-1]
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[m][n]

# Test
s1 = "abcde"
s2 = "ace"
print("Length of LCS:", longest_common_subsequence(s1, s2))
```

Task 3: 0/1 Knapsack Problem

```python
def knapsack(weights, values, capacity):
    """
    Returns the maximum value achievable with given weights and capacity.
    """
    n = len(weights)
    dp = [[0]*(capacity+1) for _ in range(n+1)]

    for i in range(1, n+1):
        for w in range(1, capacity+1):
            if weights[i-1] <= w:
                dp[i][w] = max(values[i-1] + dp[i-1][w - weights[i-1]], dp[i-1][w])
            else:
                dp[i][w] = dp[i-1][w]

    return dp[n][capacity]

# Test
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print("Maximum knapsack value:", knapsack(weights, values, capacity))
```
